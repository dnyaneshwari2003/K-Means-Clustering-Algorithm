7B. write user defined function covarience1, and calculate the value of covariance in above three cases

```python
def mean1(data):
  m = sum(data)/len(data)
  return m
def standardev(data):
  cov = 0
  resultx = mean1(data)
  for i in range(len(data)):
    cov = cov + (data[i]-resultx)*(data[i]-resultx)
  return (cov/(len(data)-1))**(0.5)
def covariance1(x,y):
  cov = 0
  resultx = mean1(x)
  resulty = mean1(y)
  for i in range(len(x)):
    cov = cov + (x[i]-resultx)*(y[i]-resulty)
  return cov/(len(x)-1)
```

```python
print(covariance1(datax,datay))
print(covariance1(datac,datad))
print(covariance1(data1,data2))
```

Explanation: These functions (mean1, standardev, and covariance1) are used to calculate the mean, standard deviation, and covariance, respectively. The calculated covariance values for the three cases (positive, negative, and no covariance) are printed.
